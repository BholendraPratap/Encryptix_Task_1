##### Feature engineering:

##### Prepare the dataset for modeling by encoding categorical variables, scaling numerical features, and handling missing values.

```python
# Define features (X) and target (y) for the training dataset
X_train = df.drop('is_fraud', axis=1)  # Features (all columns except 'is_fraud')
y_train = df['is_fraud']  # Target variable ('is_fraud')

# Define features (X) and target (y) for the testing dataset
X_test = dft.drop('is_fraud', axis=1)  # Features (all columns except 'is_fraud')
y_test = dft['is_fraud']  # Target variable ('is_fraud')

# Dynamically identify categorical and numerical columns
categorical_cols = X_train.select_dtypes(include=['object']).columns.tolist()
numerical_cols=X_train.select_dtypes(include=[np.number]).columns.tolist()
```

```python
categorical_cols
```

```python
numerical_cols
```

```python
from sklearn.preprocessing import StandardScaler,LabelEncoder
```

```python
# Encode categorical variables using LabelEncoder
label_encoder = LabelEncoder()
for col in categorical_cols:
    # Fit LabelEncoder on the union of train and test data to avoid unseen labels
    combined = pd.concat([X_train[col], X_test[col]])
    label_encoder.fit(combined)

    # Transform both train and test data
    X_train[col] = label_encoder.transform(X_train[col])
    X_test[col] = label_encoder.transform(X_test[col])
```

```python
# Scale numerical features using StandardScaler
scaler = StandardScaler()
X_train[numerical_cols] = scaler.fit_transform(X_train[numerical_cols])
X_test[numerical_cols] = scaler.transform(X_test[numerical_cols])  # Use transform (not fit_transform) for test data
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
# Ensure X_train and X_test have the same columns
X_train, X_test = X_train.align(X_test, join='inner', axis=1)
```

```python
# Print the final set of columns to verify
print("Final set of columns in X_train and X_test:")
print(X_train.columns)
print(X_test.columns)
```

#### Training the model:


##### Firstly training model without balancing the dataset:

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# Initialize the Random Forest Classifier with optimized parameters
model = RandomForestClassifier(
    n_estimators=100,       # Reduce the number of trees
    n_jobs=-1,             # Use all available CPU cores
    random_state=42
)

# Train the model on the training data
model.fit(X_train, y_train)

# Make predictions on the test data
y_pred = model.predict(X_test)
```

#### Evalutaing the model:

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Accuracy score
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

Its displaying 100% accuracy and its reflected that the model is performing for only one side not for both side and we have to need to balance the dataset. 
##### Now balancing the dataset using SMOTE technique:

```python
from imblearn.over_sampling import SMOTE
```

```python
# Apply SMOTE to balance the dataset
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)

```

```python
y_resampled.value_counts()
```

######  now our data is balanced..

```python
# Train the model on the resampled data
model.fit(X_resampled, y_resampled)

# Make predictions on the test data
y_pred_resampled = model.predict(X_test)

# Evaluate the model again
print("\nClassification Report After SMOTE:")
print(classification_report(y_test, y_pred_resampled))
```

```python
# Extract feature importance from the trained model
feature_importance = pd.DataFrame({
    'Feature': X_resampled.columns,
    'Importance': model.feature_importances_
}).sort_values(by='Importance', ascending=False)

# Print feature importance
print("\nFeature Importance:")
print(feature_importance)

# Plot feature importance
plt.figure(figsize=(8, 6))
sns.barplot(x='Importance', y='Feature', data=feature_importance)
plt.title('Feature importance')
```

##### Final evaluation curve:


```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_auc_score, roc_curve

# Assuming you have trained multiple models (e.g., Logistic Regression, Decision Tree)
# You can store them in a dictionary for easy reference.
models = {
    "Logistic Regression": log_reg,  
    "Decision Tree": dt,              
    "Random Forest": rf_model,                         
}

# Create a figure for the ROC curves
plt.figure(figsize=(8, 6))

# Loop through each model and calculate its ROC-AUC score and plot the ROC curve
for model_name, model in models.items():
    # Get model predictions (probabilities for fraud, class 1)
    y_pred_prob = model.predict_proba(X_test)[:, 1]
    
    # Calculate ROC-AUC score
    roc_auc = roc_auc_score(y_test, y_pred_prob)
    
    # Calculate ROC curve (FPR, TPR)
    fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
    
    # Plot the ROC curve for each model
    plt.plot(fpr, tpr, label=f'{model_name} (AUC = {roc_auc:.2f})')

# Plot a diagonal line for random guessing
plt.plot([0, 1], [0, 1], color='gray', linestyle='--', label="Random Guessing (AUC = 0.5)")

# Add title and labels
plt.title('Receiver Operating Characteristic (ROC) Curve Comparison')
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')

# Display legend
plt.legend(loc='lower right')

# Show the plot
plt.show()
```

##### hence both decision tree and random forest have more near to 1 which is reflecting best performance of these models.
